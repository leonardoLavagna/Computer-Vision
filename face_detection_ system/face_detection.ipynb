## Face detection system

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

%matplotlib inline
```

```python
def display(img, cmap="gray"):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
me = cv2.imread("me_con_cornetto.jpg", 0)
solvay = cv2.imread("Solvay_conference_1927.jpg", 0)
```

```python
display(me)
```

```python
display(solvay)
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
def detect_face(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img)

    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255), 10)
        
    return face_img
```

```python
result = detect_face(me)
result2 = detect_face(solvay)
```

```python
display(result)
```

```python
display(result2)
```

```python
# More accurate
def detect_face(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=1)

    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255), 10)
        
    return face_img
```

```python
result3 = detect_face(solvay)
```
